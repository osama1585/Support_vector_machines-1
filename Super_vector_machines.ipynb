# Q1. What is the mathematical formula for a linear SVM?

# Ans-

# Linear Support Vector Machine (SVM)

## Objective
Find the hyperplane that best separates the data points of two classes.

## Hyperplane Equation
The equation of the hyperplane can be written as:
\[ \mathbf{w} \cdot \mathbf{x} + b = 0 \]
where:
- \(\mathbf{w}\) is the weight vector (normal to the hyperplane),
- \(\mathbf{x}\) is the input vector,
- \(b\) is the bias term.

## Decision Rule
For a given input \(\mathbf{x}\), the classification decision is made based on:
\[ f(\mathbf{x}) = \text{sign}(\mathbf{w} \cdot \mathbf{x} + b) \]
where \(\text{sign}(z)\) returns +1 if \(z > 0\) and -1 if \(z < 0\).

## Optimization Problem
To find the optimal hyperplane, we need to solve the following optimization problem:

### Minimize:
\[ \frac{1}{2} \|\mathbf{w}\|^2 \]

subject to the constraints for all training samples \((\mathbf{x}_i, y_i)\):
\[ y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \]

Here, \(y_i \in \{-1, +1\}\) represents the class labels.

## Primal Form
The primal form of the optimization problem is:
\[ \min_{\mathbf{w}, b} \ \frac{1}{2} \|\mathbf{w}\|^2 \]

subject to:
\[ y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \quad \forall \ i \]

## Dual Form
The dual form of the optimization problem is:
\[ \max_{\alpha} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) \]

subject to:
\[ \sum_{i=1}^n \alpha_i y_i = 0 \]
\[ \alpha_i \geq 0 \quad \forall \ i \]

where \(\alpha_i\) are the Lagrange multipliers.

## Support Vectors
The solution \(\mathbf{w}\) can be expressed as a linear combination of the training samples:
\[ \mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i \]

Only the data points \(\mathbf{x}_i\) with \(\alpha_i > 0\) are support vectors and lie on the margin boundaries.

## Bias Term
The bias term \(b\) can be computed using any support vector \((\mathbf{x}_k, y_k)\):
\[ b = y_k - \mathbf{w} \cdot \mathbf{x}_k \]

In summary, a linear SVM aims to find the hyperplane that maximizes the margin between two classes by solving a convex quadratic optimization problem, either in its primal or dual form. The final classifier is determined by the weight vector \(\mathbf{w}\) and the bias term \(b\).


# Q2. What is the objective function of a linear SVM?

# Ans-

# Objective Function of a Linear SVM

The objective of a linear SVM is to find the hyperplane that maximizes the margin between two classes. This is achieved by solving the following optimization problem.

## Primal Form

### Objective:
Minimize:
\[ \frac{1}{2} \|\mathbf{w}\|^2 \]

### Subject to:
\[ y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \quad \forall \ i \]

where:
- \(\mathbf{w}\) is the weight vector,
- \(b\) is the bias term,
- \((\mathbf{x}_i, y_i)\) are the training samples with \(y_i \in \{-1, +1\}\).

## Dual Form

### Objective:
Maximize:
\[ \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j (\mathbf{x}_i \cdot \mathbf{x}_j) \]

### Subject to:
\[ \sum_{i=1}^n \alpha_i y_i = 0 \]
\[ \alpha_i \geq 0 \quad \forall \ i \]

where \(\alpha_i\) are the Lagrange multipliers.

## Summary
In both forms, the goal is to optimize the margin between the two classes, either by minimizing the norm of the weight vector (primal form) or by maximizing a function of the Lagrange multipliers (dual form), subject to the given constraints.


# Q3. What is the kernel trick in SVM?

# Ans-

# The Kernel Trick in SVM

## Introduction
The kernel trick is a technique used in Support Vector Machines (SVMs) to transform data into a higher-dimensional space to make it easier to find a hyperplane that separates the classes.

## Motivation
In some cases, data is not linearly separable in its original feature space. To address this, SVMs can use the kernel trick to implicitly map the data to a higher-dimensional space where it becomes linearly separable.

## Kernel Function
A kernel function \( K(\mathbf{x}_i, \mathbf{x}_j) \) computes the dot product of two vectors \(\mathbf{x}_i\) and \(\mathbf{x}_j\) in the transformed feature space without explicitly performing the transformation. This allows efficient computation and avoids the computational cost of working in high-dimensional spaces.

### Common Kernel Functions

1. **Linear Kernel**:
\[ K(\mathbf{x}_i, \mathbf{x}_j) = \mathbf{x}_i \cdot \mathbf{x}_j \]

2. **Polynomial Kernel**:
\[ K(\mathbf{x}_i, \mathbf{x}_j) = (\mathbf{x}_i \cdot \mathbf{x}_j + c)^d \]
   - \(c\) is a constant.
   - \(d\) is the degree of the polynomial.

3. **Radial Basis Function (RBF) / Gaussian Kernel**:
\[ K(\mathbf{x}_i, \mathbf{x}_j) = \exp\left(-\frac{\|\mathbf{x}_i - \mathbf{x}_j\|^2}{2\sigma^2}\right) \]
   - \(\sigma\) is a parameter that controls the width of the Gaussian.

4. **Sigmoid Kernel**:
\[ K(\mathbf{x}_i, \mathbf{x}_j) = \tanh(\alpha \mathbf{x}_i \cdot \mathbf{x}_j + c) \]
   - \(\alpha\) and \(c\) are kernel parameters.

## SVM with Kernel Trick

The dual form of the SVM optimization problem uses the kernel function:

### Dual Form with Kernel:
\[ \max_{\alpha} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j K(\mathbf{x}_i, \mathbf{x}_j) \]

### Subject to:
\[ \sum_{i=1}^n \alpha_i y_i = 0 \]
\[ \alpha_i \geq 0 \quad \forall \ i \]

## Summary
The kernel trick allows SVMs to create non-linear decision boundaries by implicitly mapping input features into high-dimensional spaces using kernel functions. This enables SVMs to solve complex classification problems that are not linearly separable in the original feature space.


# Q4. What is the role of support vectors in SVM Explain with example

# Role of Support Vectors in SVM

## Introduction
Support Vectors are the key elements of the training dataset used by the Support Vector Machine (SVM) to define the decision boundary that separates the classes. They are the data points that lie closest to the decision boundary and directly influence its position and orientation.

## Decision Boundary
In an SVM, the decision boundary is determined by a hyperplane, which is defined as:
\[ \mathbf{w} \cdot \mathbf{x} + b = 0 \]
where:
- \(\mathbf{w}\) is the weight vector,
- \(\mathbf{x}\) is the input vector,
- \(b\) is the bias term.

## Margin
The margin is the distance between the hyperplane and the nearest data points from either class. SVM aims to maximize this margin. The data points that lie exactly on the margin are the support vectors.

## Example
Consider a 2-dimensional dataset with two classes (red and blue):

- Red class: points \((1, 2)\), \((2, 3)\), \((3, 3)\)
- Blue class: points \((6, 6)\), \((7, 8)\), \((8, 6)\)

Suppose the SVM finds the following hyperplane:
\[ \mathbf{w} \cdot \mathbf{x} + b = 0 \]

### Support Vectors
In this example, let's say the support vectors are:
- \((2, 3)\) from the red class,
- \((6, 6)\) from the blue class.

These points are closest to the hyperplane and lie on the margin.

# Illustrations of Hyperplane, Marginal Plane, Soft Margin, and Hard Margin in SVM

## 1. Hyperplane
The hyperplane is the decision boundary that separates the classes. In a 2-dimensional space, it is a line, while in higher dimensions, it is a plane or a hyperplane.

For a 2-dimensional dataset, the hyperplane can be represented as:
\[ \mathbf{w} \cdot \mathbf{x} + b = 0 \]

### Graph:



```python

import matplotlib.pyplot as plt
import numpy as np

# Plotting the hyperplane
x = np.linspace(0, 10, 100)
y = -x + 5

plt.plot(x, y, '-r', label='Hyperplane: $w \cdot x + b = 0$')
plt.scatter([2, 3, 6, 7], [3, 2, 7, 6], color='blue')  # Example points
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Hyperplane in SVM')
plt.show()
```

# MARGINAL PLANE:

```python
# Plotting the hyperplane and marginal planes
y_margin1 = -x + 6
y_margin2 = -x + 4

plt.plot(x, y, '-r', label='Hyperplane: $w \cdot x + b = 0$')
plt.plot(x, y_margin1, '--g', label='Margin: $w \cdot x + b = 1$')
plt.plot(x, y_margin2, '--g', label='Margin: $w \cdot x + b = -1$')
plt.scatter([2, 3, 6, 7], [3, 2, 7, 6], color='blue')  # Example points
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Hyperplane and Marginal Planes in SVM')
plt.show()
```

# HARD MARGIN:

```python
# Plotting the hard margin SVM
plt.plot(x, y, '-r', label='Hyperplane: $w \cdot x + b = 0$')
plt.plot(x, y_margin1, '--g', label='Margin: $w \cdot x + b = 1$')
plt.plot(x, y_margin2, '--g', label='Margin: $w \cdot x + b = -1$')
plt.scatter([2, 3], [3, 2], color='blue', label='Class 1')  # Class 1 points
plt.scatter([6, 7], [7, 6], color='red', label='Class 2')   # Class 2 points
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Hard Margin SVM')
plt.show()
```

# SOFT MARGIN:

```python
# Plotting the soft margin SVM
plt.plot(x, y, '-r', label='Hyperplane: $w \cdot x + b = 0$')
plt.plot(x, y_margin1, '--g', label='Margin: $w \cdot x + b = 1$')
plt.plot(x, y_margin2, '--g', label='Margin: $w \cdot x + b = -1$')
plt.scatter([2, 3, 5, 6], [3, 2, 4, 7], color='blue', label='Class 1')  # Class 1 points
plt.scatter([7, 8, 5], [6, 5, 5], color='red', label='Class 2')   # Class 2 points
plt.scatter([5], [5], color='purple', label='Misclassified')   # Misclassified point
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Soft Margin SVM')
plt.show()
```

```python
# SVM Implementation through Iris Dataset

## 1. Load the Iris Dataset and Split it into Training and Testing Sets

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # We take the first two features for visualization purposes
y = iris.target

# Binary classification: Only take the first two classes
X = X[y != 2]
y = y[y != 2]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset (first two features)')
plt.show()
```

# # Train a linear SVM classifier

```python

clf = SVC(kernel='linear', C=1.0)
clf.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)

# Compute the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of scikit-learn SVM: {accuracy:.2f}')
```

# Plot the decision boundaries of the train model

```python
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Set1)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Set1)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title('Decision Boundary')
    plt.show()

plot_decision_boundary(clf, X, y)
```

# Train the different value of the regularization parameter C

```python
C_values = [0.1, 1, 10, 100]
for C in C_values:
    clf = SVC(kernel='linear', C=C)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy with C={C}: {accuracy:.2f}')
    plot_decision_boundary(clf, X, y)
```

# Bonus Task

```python
class LinearSVM:
    def __init__(self, C=1.0, lr=0.001, n_iters=1000):
        self.C = C
        self.lr = lr
        self.n_iters = n_iters

    def fit(self, X, y):
        n_samples, n_features = X.shape
        y_ = np.where(y <= 0, -1, 1)  # Convert y to -1, 1

        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.C * self.w)
                else:
                    self.w -= self.lr * (2 * self.C * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.lr * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) - self.b
        return np.sign(approx)

# Train the linear SVM classifier from scratch
svm = LinearSVM(C=1.0, lr=0.001, n_iters=1000)
svm.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred_scratch = svm.predict(X_test)

# Compute the accuracy of the model
accuracy_scratch = accuracy_score(y_test, y_pred_scratch)
print(f'Accuracy of custom SVM: {accuracy_scratch:.2f}')

# Plot the decision boundary
plot_decision_boundary(svm, X, y)
```
